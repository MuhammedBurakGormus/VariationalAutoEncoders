**Examining the effect of number of layer**

In this notebook, the losses coming from different VAEs architectures are plotted using their pickle files. 

```python
import pickle
#Â Get the losses
with open('2Layers/architecture_loss_history_2_layers.pkl', 'rb') as f:
    architecture_loss_history_2_layers = pickle.load(f)
with open('3Layers/architecture_loss_history_3_layers.pkl', 'rb') as f:
    architecture_loss_history_3_layers = pickle.load(f)
with open('4Layers/architecture_loss_history_4_layers.pkl', 'rb') as f:
    architecture_loss_history_4_layers = pickle.load(f)
```

```python
import matplotlib.pyplot as plt
# Create x values from 1 to the length of loss_data
x = list(range(1, len(architecture_loss_history_4_layers) + 1))
# Plot the lines
plt.plot(x,architecture_loss_history_2_layers,'-',label="2Layers")
plt.plot(x,architecture_loss_history_3_layers,'-',label="3Layers")
plt.plot(x,architecture_loss_history_4_layers,'-',label="4Layers")
# Add labels and title
plt.xlabel('Epochs')
plt.ylabel('Average loss')
plt.title('The effect of number of layers in average loss')
# Set the legend
plt.legend()
# Add a grid
plt.grid()
# Save the plot
plt.savefig("The_effect_of_number_of_layers_in_average_loss")
```
